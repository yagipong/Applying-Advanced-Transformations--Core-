```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json
```

```python
df_info = pd.read_csv('Data/superhero_info.csv')
df_info.head()
```

```python
df_info[['Hero', 'Publisher']] = df_info['Hero|Publisher'].str.split('|', expand = True)
df_info.head()
```

```python
df_info = df_info.drop(columns = 'Hero|Publisher')
df_info.head()
```

```python
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
measurements = df_info['Measurements'].apply(pd.Series)

df_info = pd.concat((df_info, measurements), axis = 1)
df_info.head()
```

```python
df_info = df_info.drop(columns = 'Measurements')
```

```python
df_info.head()
```

```python
df_info['Height'] = df_info['Height'].str.replace(" cm", '')
df_info['Height'] = df_info['Height'].apply(json.loads)
df_info['Height'].head()
```

```python
df_info['Weight'] = df_info['Weight'].str.replace(" kg", '')
df_info['Weight'] = df_info['Weight'].apply(json.loads)
df_info['Weight'].head()
```

```python
df_info.head()
```

```python
df_info = df_info[['Hero', 'Publisher', 'Eye color', 'Race', 'Hair color', 'Height', 
         'Skin color', 'Alignment', 'Weight']]
df_info.head()
```

```python
df_powers = pd.read_csv('Data/superhero_powers.csv')
df_powers.head()
```

```python
df_powers['Powers_Unique'] = df_powers['Powers'].str.split(',')
df_powers['Powers_Unique']
```

```python
exploded = df_powers.explode('Powers_Unique')
exploded
```

```python
cols_to_make = exploded['Powers_Unique'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df_powers[col] = df_powers['Powers'].str.contains(col);
    
df_powers.head()
```

```python
df_powers = df_powers.drop(columns = ['Powers_Unique', 'Powers'])
df_powers.head()
```

```python
merged = pd.merge(df_info, df_powers, left_on = 'Hero', right_on = 'hero_names', how = 'inner')
merged
```

Question 1

```python
super_speed_yes = merged['Super Speed'] == True
super_speed_no  = merged['Super Speed'] == False
```

```python
df_yes = merged.loc[super_speed_yes]
fast_weight = df_yes[['Weight']].mean()
fast_weight
```

```python
df_no = merged.loc[super_speed_no]
slow_weight = df_no[['Weight']].mean()
slow_weight
```

```python
diff = fast_weight - slow_weight
diff
```
